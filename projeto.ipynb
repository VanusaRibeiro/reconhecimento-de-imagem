```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
from skimage.io import imread
from IPython.core.display import Image



sinasc_jan = pd.read_csv('/content/input/SINASC_RO_2019_JAN.csv')
sinasc_jan.head()


sinasc_mar = pd.read_csv('/content/input/SINASC_RO_2019_MAR.csv')
sinasc_mar.head()

sinasc_abr = pd.read_csv('/content/input/SINASC_RO_2019_ABR.csv')
sinasc_abr.head()

sinasc_mai = pd.read_csv('/content/input/SINASC_RO_2019_MAI.csv')
sinasc_mai.head()

sinasc_jun = pd.read_csv('/content/input/SINASC_RO_2019_JUN.csv')
sinasc_jun.head()

sinasc_dez = pd.read_csv('/content/input/SINASC_RO_2019_DEZ.csv')
sinasc_jun.head()

def plota_pivot_table(df, value, index, func, ylabel, xlabel, opcao='nada'):
    if opcao == 'nada':
        pd.pivot_table(df, values=value, index=index,
                       aggfunc=func).plot(figsize=[15, 5])
    elif opcao == 'sort':
        pd.pivot_table(df, values=value, index=index,
                       aggfunc=func).sort_values(value).plot(figsize=[15, 5])
    elif opcao == 'unstack':
        pd.pivot_table(df, values=value, index=index,
                       aggfunc=func).unstack().plot(figsize=[15, 5])
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    return None

max_data_jan = (sinasc_jan.DTNASC.max()[:7])
print(max_data_jan)
max_data_mar = (sinasc_mar.DTNASC.max()[:7])
print(max_data_mar)
max_data_abr = (sinasc_abr.DTNASC.max()[:7])
print(max_data_abr)
max_data_mai = (sinasc_mai.DTNASC.max()[:7])
print(max_data_mai)
max_data_jun = (sinasc_jun.DTNASC.max()[:7])
print(max_data_jun)
max_data_dez = (sinasc_dez.DTNASC.max()[:7])
print(max_data_dez)
max_data_1 = (sinasc_dez.DTNASC.max()[:7])
print(max_data_1)

os.makedirs('./input/figs/2019-12'+max_data_jan, exist_ok=True)

os.makedirs('./input/figs/2019-12'+max_data_mar, exist_ok=True)

os.makedirs('./input/figs/2019-12'+max_data_abr, exist_ok=True)

os.makedirs('./input/figs/2019-12'+max_data_mai, exist_ok=True)

os.makedirs('./input/figs/2019-12'+max_data_jun, exist_ok=True)

os.makedirs('./input/figs/2019-12'+max_data_dez, exist_ok=True)

#Relatório do mes de Janeiro

plota_pivot_table(sinasc_jan, 'IDADEMAE', ['DTNASC', 'SEXO'], 'mean', 'media idade mae','data de nascimento','unstack')
plt.savefig('./input/figs/2019-12'+max_data_jan+'/media idade mae por sexo.png')


plota_pivot_table(sinasc_jan, 'PESO', ['DTNASC', 'SEXO'], 'mean', 'media peso bebe','data de nascimento','unstack')
plt.savefig('./input/figs/2019-12'+max_data_jan+'/media peso bebe por sexo.png')


plota_pivot_table(sinasc_jan, 'PESO', 'ESCMAE', 'median', 'apgar1 medio','gestacao','sort')
plt.savefig('./input/figs/2019-12'+max_data_jan+'/media apgar1 por escolaridade mae.png')

plota_pivot_table(sinasc_abr, 'APGAR1', 'GESTACAO', 'mean', 'apgar1 medio','gestacao','sort')
plt.savefig('./input/figs/2019-12'+max_data_jan+'/media apgar1 por gestacao.png')


plota_pivot_table(sinasc_jan, 'APGAR5', 'GESTACAO', 'mean', 'apgar5 medio','gestacao','sort')
plt.savefig('./input/figs/2019-12'+max_data_jan+'/media apgar5 por gestacao.png')

# Relatório do mês de março

plota_pivot_table(sinasc_mar, 'IDADEMAE', ['DTNASC', 'SEXO'], 'mean', 'media idade mae','data de nascimento','unstack')
plt.savefig('./input/figs/2019-12'+max_data_mar+'/media idade mae por sexo.png')

plota_pivot_table(sinasc_mar, 'PESO', ['DTNASC', 'SEXO'], 'mean', 'media peso bebe','data de nascimento','unstack')
plt.savefig('./input/figs/2019-12'+max_data_mar+'/media peso bebe por sexo.png')

plota_pivot_table(sinasc_mar, 'PESO', 'ESCMAE', 'median', 'apgar1 medio','gestacao','sort')
plt.savefig('./input/figs/2019-12'+max_data_mar+'/media apgar1 por escolaridade mae.png')

plota_pivot_table(sinasc_abr, 'APGAR1', 'GESTACAO', 'mean', 'apgar1 medio','gestacao','sort')
plt.savefig('./input/figs/2019-12'+max_data_mar+'/media apgar1 por gestacao.png')

plota_pivot_table(sinasc_mar, 'APGAR1', 'GESTACAO', 'mean', 'apgar1 medio','gestacao','sort')
plt.savefig('./input/figs/2019-12'+max_data_mar+'/media apgar5 por gestacao.png')

#Relatório do mês de Abril

plota_pivot_table(sinasc_abr, 'IDADEMAE', ['DTNASC', 'SEXO'], 'mean', 'media idade mae','data de nascimento','unstack')
plt.savefig('./input/figs/2019-12'+max_data_abr+'/media idade mae por sexo.png')


plota_pivot_table(sinasc_abr, 'PESO', ['DTNASC', 'SEXO'], 'mean', 'media peso bebe','data de nascimento','unstack')
plt.savefig('./input/figs/2019-12'+max_data_abr+'/media peso bebe por sexo.png')

plota_pivot_table(sinasc_abr, 'PESO', 'ESCMAE', 'median', 'apgar1 medio','gestacao','sort')
plt.savefig('./input/figs/2019-12'+max_data_abr+'/media apgar1 por escolaridade mae.png')

plota_pivot_table(sinasc_abr, 'APGAR1', 'GESTACAO', 'mean', 'apgar1 medio','gestacao','sort')
plt.savefig('./input/figs/2019-12'+max_data_abr+'/media apgar1 por gestacao.png')

plota_pivot_table(sinasc_abr, 'APGAR1', 'GESTACAO', 'mean', 'apgar1 medio','gestacao','sort')
plt.savefig('./input/figs/2019-12'+max_data_abr+'/media apgar5 por gestacao.png')


#Relatório do mês de maio


plota_pivot_table(sinasc_mai, 'IDADEMAE', ['DTNASC', 'SEXO'], 'mean', 'media idade mae','data de nascimento','unstack')
plt.savefig('./input/figs/2019-12'+max_data_mai+'/media idade mae por sexo.png')

plota_pivot_table(sinasc_mai, 'PESO', ['DTNASC', 'SEXO'], 'mean', 'media peso bebe','data de nascimento','unstack')
plt.savefig('./input/figs/2019-12'+max_data_mai+'/media peso bebe por sexo.png')

plota_pivot_table(sinasc_mai, 'PESO', 'ESCMAE', 'median', 'apgar1 medio','gestacao','sort')
plt.savefig('./input/figs/2019-12'+max_data_mai+'/media apgar1 por escolaridade mae.png')

plota_pivot_table(sinasc_mai, 'APGAR1', 'GESTACAO', 'mean', 'apgar1 medio','gestacao','sort')
plt.savefig('./input/figs/2019-12'+max_data_mai+'/media apgar1 por gestacao.png')

plota_pivot_table(sinasc_mai, 'APGAR1', 'GESTACAO', 'mean', 'apgar1 medio','gestacao','sort')
plt.savefig('./input/figs/2019-12'+max_data_mai+'/media apgar1 por gestacao.png')


#Relatório do mês de junho


plota_pivot_table(sinasc_jun, 'IDADEMAE', ['DTNASC', 'SEXO'], 'mean', 'media idade mae','data de nascimento','unstack')
plt.savefig('./input/figs/2019-12'+max_data_jun+'/media idade mae por sexo.png')

plota_pivot_table(sinasc_jun, 'PESO', ['DTNASC', 'SEXO'], 'mean', 'media peso bebe','data de nascimento','unstack')
plt.savefig('./input/figs/2019-12'+max_data_jun+'/media peso bebe por sexo.png')

plota_pivot_table(sinasc_jun, 'PESO', 'ESCMAE', 'median', 'apgar1 medio','gestacao','sort')
plt.savefig('./input/figs/2019-12'+max_data_jun+'/media apgar1 por escolaridade mae.png')

plota_pivot_table(sinasc_jun, 'APGAR1', 'GESTACAO', 'mean', 'apgar1 medio','gestacao','sort')
plt.savefig('./input/figs/2019-12'+max_data_jun+'/media apgar1 por gestacao.png')

plota_pivot_table(sinasc_jun, 'APGAR1', 'GESTACAO', 'mean', 'apgar1 medio','gestacao','sort')
plt.savefig('./input/figs/2019-12'+max_data_jun+'/media apgar1 por gestacao.png')





#Relatório do mes de dezembro

plota_pivot_table(sinasc_dez, 'IDADEMAE', ['DTNASC', 'SEXO'], 'mean', 'media idade mae','data de nascimento','unstack')
plt.savefig('./input/figs/2019-12'+max_data_dez+'/media idade mae por sexo.png')


plota_pivot_table(sinasc_dez, 'PESO', ['DTNASC', 'SEXO'], 'mean', 'media peso bebe','data de nascimento','unstack')
plt.savefig('./input/figs/2019-12'+max_data_dez+'/media peso bebe por sexo.png')

plota_pivot_table(sinasc_dez, 'PESO', 'ESCMAE', 'median', 'apgar1 medio','gestacao','sort')
plt.savefig('./input/figs/2019-12'+max_data_dez+'/media apgar1 por escolaridade mae.png')

plota_pivot_table(sinasc_dez, 'APGAR1', 'GESTACAO', 'mean', 'apgar1 medio','gestacao','sort')
plt.savefig('./input/figs/2019-12'+max_data_dez+'/media apgar1 por gestacao.png')

plota_pivot_table(sinasc_dez, 'APGAR1', 'GESTACAO', 'mean', 'apgar1 medio','gestacao','sort')
plt.savefig('./input/figs/2019-12'+max_data_dez+'/media apgar1 por gestacao.png')


```
